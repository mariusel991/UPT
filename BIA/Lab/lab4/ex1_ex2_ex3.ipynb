```python
import numpy as np
import os

np.random.seed(42)

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeRegressor
from graphviz import Source
from sklearn.tree import export_graphviz

X = np.random.rand(350,1)
y = (X + 1.2)**3 - 10 + np.random.rand(350,1)/10

tree_reg = DecisionTreeRegressor(max_depth=3, min_samples_leaf=5, random_state=42)
tree_reg.fit(X, y)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
def plot_regression_predictions(tree_reg, X, y, axes):
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    y_pred = tree_reg.predict(x1)
    plt.axis(axes)
    plt.xlabel("$x_1$", fontsize=18)
    plt.ylabel("$y$", fontsize=18, rotation=0)
    plt.plot(X, y, "b.")
    plt.plot(x1, y_pred, "r.-", linewidth=2, label=r"$\hat{y}$")


fig, ax = plt.subplots(figsize=(10, 4))  # Removed ncols=2

plot_regression_predictions(tree_reg, X, y, axes=[0.5, 1, -5, 1])
plt.legend(loc="upper left", fontsize=18)
plt.title("max_depth=5", fontsize=14)


plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
digits = load_digits()

X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(),
    n_estimators=200,
    max_samples=100,
    bootstrap=True,
    random_state=42)

bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

```python
print("Bagging classifier: ", accuracy_score(y_test, y_pred))
```
